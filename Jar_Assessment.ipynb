# *Part A Walmart Analysis*

```python
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

df = pd.read_excel("C:\\Users\\HP\\OneDrive\\Desktop\\Walmart_Sales.xlsx")
```

```python
df.describe()
```

```python
df.info
```

```python
df['Revenue_Generated'] = (df['Unit price'] * df['Quantity'])
```

```python
df['Revenue_Generated']
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
```

```python
df['Date']
```

```python
df['Month'] = pd.to_datetime(df['Date']).dt.month 
df['Year'] = pd.to_datetime(df['Date']).dt.year
```

```python
df
```

## A. Analyze the performance of sales and revenue at the city and branch level

```python
Sales_by_City_Branch = df.groupby(['City', 'Branch'])[['Quantity', 'Revenue_Generated']].sum()
print("Sales and Revenue at City and Branch Level:")
print(Sales_by_City_Branch)
```

## B. What is the average price of an item sold at each branch of the city


```python
Avg_Price_by_Branch_City = df.groupby(['City', 'Branch'])['Unit price'].mean() 
print("\nAverage Price of an Item Sold at Each Branch of the City:") 
print(Avg_Price_by_Branch_City)
```

## C. Analyze the performance of sales and revenue, Month over Month across the Product line, Gender, and Payment Method

```python
Analyzed_data = df.groupby(['Year', 'Month', 'Product line', 'Gender', 'Payment'])[['Quantity','Revenue_Generated']].sum()
```

```python
Analyzed_data
```

## D. Calculate month-over-month change in sales quantity and revenue


```python
Analyzed_data['Sales_Quantity_MoM_Change'] = Analyzed_data.groupby(['Product line', 'Gender'])['Quantity'].pct_change()
```

```python
Analyzed_data['Revenue_MoM_Change'] = Analyzed_data.groupby(['Product line', 'Gender'])['Revenue_Generated'].pct_change()
```

```python
Analyzed_data
```

## E. Identify focus areas for improvement based on trends

```python
focus_areas = Analyzed_data.groupby(['Product line', 'Gender', 'Payment'])[['Sales_Quantity_MoM_Change', 'Revenue_MoM_Change']].mean().reset_index()
```

```python
focus_areas = focus_areas[(focus_areas['Sales_Quantity_MoM_Change'] > 0) | (focus_areas['Revenue_MoM_Change'] > 0)]
```

```python
print("Focus Areas with Positive Month-over-Month Changes:")
print(focus_areas)
```

```python
plt.figure(figsize=(14, 8))

# Sales Quantity MoM Change Plot
plt.subplot(1, 2, 1)
barplot1 = sns.barplot(data=focus_areas, x='Product line', y='Sales_Quantity_MoM_Change', hue='Gender', dodge=True)
plt.title('Sales Quantity MoM Change by Product Line and Gender')
plt.xticks(rotation=45)
plt.xlabel('Product Line')
plt.ylabel('Sales Quantity MoM Change')
plt.legend(title='Gender')

# Add annotations
for p in barplot1.patches:
    barplot1.annotate(format(p.get_height(), '.2f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center', va = 'center',
                      xytext = (0, 9),
                      textcoords = 'offset points')

# Revenue MoM Change Plot
plt.subplot(1, 2, 2)
barplot2 = sns.barplot(data=focus_areas, x='Product line', y='Revenue_MoM_Change', hue='Gender', dodge=True)
plt.title('Revenue MoM Change by Product Line and Gender')
plt.xticks(rotation=45)
plt.xlabel('Product Line')
plt.ylabel('Revenue MoM Change')
plt.legend(title='Gender')

# Add annotations
for p in barplot2.patches:
    barplot2.annotate(format(p.get_height(), '.2f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center', va = 'center',
                      xytext = (0, 9),
                      textcoords = 'offset points')

# Adjust the layout
plt.tight_layout()
plt.show()
```

## Explaination:

#### -Bar Plots: Two side-by-side bar plots are created using seaborn. Bar plots are chosen because they are effective in comparing quantities across different categories.
#### -First Plot (Sales Quantity MoM Change): This plot shows the Month-over-Month change in sales quantity for each product line, broken down by gender.
#### -Second Plot (Revenue MoM Change): This plot shows the Month-over-Month change in revenue for each product line, also broken down by gender.
#### -Tight Layout: plt.tight_layout() is used to ensure that the subplots do not overlap and there is adequate spacing between them. This makes the visualization more readable and aesthetically pleasing.


#### -Highlight Positive Trends: The visualization aims to highlight areas where there are positive Month-over-Month changes in sales quantity or revenue. This is important for identifying which product lines, genders, and payment methods are contributing to growth.
#### -Identify Performance Drivers: By breaking down the data by product line, gender, and payment method, the visualization helps in pinpointing specific segments that are performing well.


### Key Insights

#### -Product Line Performance: You can quickly identify which product lines have the highest positive MoM changes in sales quantity and revenue.
#### -Gender Preferences: The visualization allows for the comparison of male and female preferences and performance within each product line, which can inform targeted marketing strategies.
#### -Strategic Focus Areas: By focusing on segments with positive MoM changes, businesses can allocate resources more effectively, doubling down on successful areas and re-evaluating strategies in underperforming ones.
